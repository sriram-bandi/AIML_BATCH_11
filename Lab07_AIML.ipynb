Implement Support Vector Machine Classification using Breast Cancer Dataset

```python
  import matplotlib.pyplot as plt
  import numpy as np
  import pandas as pd
  import seaborn as sns
  %matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
c=load_breast_cancer()
```

```python
c.keys()
```

```python
print(c['DESCR'])
```

```python
c['feature_names']
```

```python
dataframe=pd.DataFrame(c['data'],columns=c['feature_names'])
dataframe.info()
```

```python
dataframe.describe()
```

```python
np.sum(pd.isnull(dataframe).sum())
```

```python
c['target']
```

```python
dataframe['c']=pd.DataFrame(c['target'])
dataframe.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='c',data=dataframe,palette='rainbow')
```

```python
list1=list(dataframe.columns[0:10])
for i in range(len(list1)-1):
  sns.boxplot(x='c',y=list1[i],data=dataframe)
  plt.figure()
```

```python
f,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(12,6))
ax1.scatter(dataframe['mean area'],dataframe['c'])
ax1.set_title('Cancer cases as a function of mean of area:',fontsize=10)
ax2.scatter(dataframe['mean smoothness'],dataframe['c'])
ax2.set_title('Cancer cases as a function of mean of smoothness:',fontsize=10)
```

```python
df_feat=dataframe.drop('c',axis=1)
df_feat.head()
```

```python
df_target=dataframe['c']
df_target.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df_feat,df_target,test_size=0.30,random_state=101)
```

```python
y_train.head()
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
print('MISCALCULATION ERROR RATE:',round(np.mean(predictions!=y_test),3))
```
